```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training dat
train_df

train = train_df.drop(columns=['Unnamed: 0', 'index', 'loan_status'])
train
```

```python
train.columns
```

```python
train.shape
```

```python
X_train = pd.get_dummies(train, dtype = float)
X_train.head()
```

```python
y_train = train_df['loan_status']
y_train.value_counts()
```

```python
test_df

test = test_df.drop(columns=['Unnamed: 0', 'index', 'loan_status'])
test
```

```python
X_test = pd.get_dummies(test, dtype = float)
X_test.head()
```

```python
y_test = test_df['loan_status']
y_test.value_counts()
```

```python
y_test.head()
```

```python
# Finding difference between X_train and X_test columns add missing dummy variables to testing set
set(list(X_train.columns))- set(list(X_test.columns))
```

```python
#make a list for for debt_settlement_flag_y
X_test['debt_settlement_flag_y'] = pd.Series ([0 for x in range (len(X_test.index))])
X_test.head()
```

My prediction is that the RandomForest model will perform better as it is more focused on accuracy and does well with categorical datasets. RandomForest is also less prone to overfitting.

#Logistic Regression Model

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train,y_train)
log_unscaled=classifier.fit(X_train, y_train)
```

```python
#print model scores
print(f'Training Score: {classifier.score(X_train, y_train)}')
print(f'Testing Score: {classifier.score(X_test, y_test)}')
```

```python
#report
from sklearn.metrics import classification_report
y_true = y_test
y_pred = classifier.predict(X_test)
print(classification_report(y_true, y_pred))
```

#Random Forest Classifier

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
classifier = RandomForestClassifier(random_state=1,n_estimators=400).fit(X_train, y_train)
```

```python
print(f'Training Score: {classifier.score(X_train, y_train)}')
print(f'Testing Score: {classifier.score(X_test, y_test)}')
```

```python
# classification Report
from sklearn.metrics import classification_report
y_true = y_test
y_pred = classifier.predict(X_test)
print(classification_report(y_true, y_pred))
```

Between the LogisticRegression and the the RandomForest Classifier models, the RandomForest had a better training and testing score than the LogisticRegression. Also when looking at their classification reports, RandomForest's f1-score was higher or more close to 1.

```python
#scaled data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

Normalizing the data is to help with optimization. It does not really affect RandomForest models. However, with LogisticRegression, normalized data makes it easier to interpret expected values. My prediction will be again that LogisticRegression model will perform better than RandomForest.

```python
# Scale the data
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier=LogisticRegression().fit(X_train_scaled, y_train)
print(f'Training Score: {classifier.score(X_train_scaled, y_train)}')
print(f'Testing Score: {classifier.score(X_test_scaled, y_test)}')
```

```python
#LogisticRegression Scaled Classification Report
from sklearn.metrics import classification_report
y_true = y_test
y_pred = classifier.predict(X_test)
print(classification_report(y_true, y_pred))
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
classifier = RandomForestClassifier(random_state=1, n_estimators=400).fit(X_train_scaled, y_train)
print(f'Training Score: {classifier.score(X_train_scaled, y_train)}')
print(f'Testing Score: {classifier.score(X_test_scaled, y_test)}')
```

```python
#RandomForest Scaled Classification Report
from sklearn.metrics import classification_report
y_true = y_test
y_pred = classifier.predict(X_test)
print(classification_report(y_true, y_pred))
```

With the normalized data, the LogisticRegression model had a decent training and testing score. The RandomForest model had a really good training score and a decent testing score, though it is lower than the LogisticRegression testing score. When evaluating the classification reports, LogisticRegression had better overall precision and f1-score.
